# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

## Basic

```python
plt.plot(x, y, '--r', linewidth=4,  marker='*', markersize=15, markeredgewidth=2, markeredgecolor='g', markerfacecolor='b')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Super Title')
plt.show()
```

## Multiplots

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*--')
plt.show()
```

## Fig

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y, 'b')
ax.set_xlabel('X title')
ax.set_ylabel('y title')
ax.set_title('Super title')
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x, y, 'b')
ax1.set_xlabel('x1 title')
ax1.set_ylabel('y1 title')

ax2.plot(y, x, 'r')
ax2.set_xlabel('x2 title')
ax2.set_ylabel('y2 title')

plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x', fontsize=22, color='red')
    ax.set_ylabel('y')
    ax.set_title('title')

fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```

## Figsize

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 5))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, label='xY')
ax.plot(y, x, label='yX')
ax.legend(loc='upper left')

plt.show()
```

## Style

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5)
ax.plot(x, x+2, color="#8B008B")
ax.plot(x, x+3, color="#FF8C00")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10])

ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")

plt.show()
```

## Plot range

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range")

plt.show()
```
